## 2: Plot 12 most dangerous cities

```python
city_freq = kill.city.value_counts().head(12)
city_freq
```

```python
# with seaborn barplot
fig, ax =plt.subplots(figsize=(8,5))

sns.barplot(x=city_freq.index, y=city_freq.values)
ax.set_title('Most dangerous cities', fontdict= { 'fontsize': 14, 'color':'b'})
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45);
```

```python
# with seaborn countplot
city_12 = kill[kill.city.isin(city_freq.index)]

fig, ax =plt.subplots(figsize=(8,5))

sns.countplot(x='city', data=city_12, order=city_freq.index)
ax.set_title('Most dangerous cities', fontdict= { 'fontsize': 14, 'color':'b'})
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45);
```
